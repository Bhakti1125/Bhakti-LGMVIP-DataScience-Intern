# Overview of the Project 

The image to pencil sketch project converts an input image into a grayscale representation, applies image inversion and Gaussian blur, and blends it with the original grayscale image to create a pencil sketch effect. The resulting sketch can be displayed or saved as an output file.

```python
#importing the required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image file
image = cv2.imread('C:/Users/BHAKTI/OneDrive/Desktop/tiger.jpeg')
```

```python
def dodgeV2(image, mask):
    return cv2.divide(image, 255 - mask, scale=256.0)
```

```python
def pencil_sketch(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Display the grayscale image
    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')
    plt.show()
    
    # Invert the grayscale image
    inverted_gray = cv2.bitwise_not(gray_image)
    # Display the inverted grayscale image
    plt.imshow(inverted_gray, cmap='gray')
    plt.title('Inverted Grayscale Image')
    plt.axis('off')
    plt.show()
    
    # Apply Gaussian blur to the inverted image
    blurred_image = cv2.GaussianBlur(inverted_gray, (21, 21), 0)
    # Display the blurred image
    plt.imshow(blurred_image, cmap='gray')
    plt.title('Blurred Image')
    plt.axis('off')
    plt.show()
    
    # Blend the grayscale image with the blurred image using dodgeV2 function
    pencil_sketch = dodgeV2(gray_image, blurred_image)
    # Display the pencil sketch image
    plt.imshow(pencil_sketch, cmap='gray')
    plt.title('Pencil Sketch')
    plt.axis('off')
    plt.show()

    return pencil_sketch
```

```python
# Create a pencil sketch
sketch = pencil_sketch(image)
```

```python
# Display the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()

# Display the pencil sketch image
plt.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()
```
